```python
all_files = glob.glob(os.path.join(filepaths, '*.csv'))
li = []
```

```python
for f in all_files:
    df = pd.read_csv(f, index_col=None, header=0)
    li.append(df)
```

```python
april_20, aug_19, dec_19, feb_20, jan_20, jul_19, jun_19, mar_20, may_20, nov_19, oct_19, sept_19  = li[0], li[1], li[2], li[3], li[4], li[5], li[6], li[7], li[8], li[9], li[10], li[11]
```

```python
jun19_may20_df = pd.concat([jun_19,jul_19,aug_19,sept_19,oct_19,nov_19,dec_19,jan_20,feb_20,mar_20,april_20,may_20], axis=0)
```

```python
jun19_may20_df.reset_index()
```

```python
jun19_may20_df['sto_time'] = jun19_may20_df['stop_time'].str.replace(' AM', '')
```

```python
jun19_may20_df.rename(columns={'trip_duration':'trip_duration_seconds'}, inplace=True)
jun19_may20_df
```

```python
trip_duration_minutes = [] 
for m in jun19_may20_df['trip_duration_seconds']:
    minutes = m / 60
    trip_duration_minutes.append(minutes)
```

```python
jun19_may20_df['trip_duration_minutes'] = trip_duration_minutes 
```

```python
jun19_may20_df
```

```python
gender = []
for g in jun19_may20_df['gender']:
    if g == 0:
        gender.append('unknown')
    elif g == 1:
        gender.append('male')
    else:
        gender.append('female')

    
```

```python
jun19_may20_df.to_csv('cleaned_citibike_data', index=False)
```
